```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("updatedJobs.csv")
time = df["Full-Time/Part-Time indicator"]
```

```python
extra = df["Salary Range From"]
```

```python
counter = 0
myArray = []
for x in time:
    if (x == None):
        myArray.append("Null")
    elif (len(x) > 3):
        myArray.append(extra[counter])
    else:
        myArray.append(x)
    counter+=1
```

```python
print(myArray)
```

```python
df['UpdatedWorkTime'] = myArray
```

```python
df.to_csv("updatedJobs3.csv")
```
