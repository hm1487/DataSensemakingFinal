```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("NYC_Jobs.csv")
```

```python
departments = df['Job Category']
print(len(departments))
```

```python
deptDict = {}
for d in departments:
    temp = d.split(" ")
    if (temp[0] == "Health"):
        holder = temp[0]
    elif (temp[0] == ""):
        holder = "Null"
    elif (temp[0] == "Legal"):
        holder = temp[0]
    elif (len(temp) > 1):
        if (temp[1] == "&"):
            holder = temp[0] + " " + temp[1] + " " + temp[2]
        else:
            holder = temp[0] + " " + temp[1]
    else:
        holder = temp[0]
    if (deptDict.get(holder)==None):
        deptDict[holder] = 1
    else:
        deptDict[holder] += 1
```

```python
for x in sorted(deptDict, key=deptDict.get, reverse=True):
    if (x == ""):
        print("Null", deptDict.get(x))
    else:
        print(x,deptDict.get(x))
total = 0
for x in deptDict.values():
    total += x
print(total)
```

```python
newDepartments = []
newDD = {}
for d in departments:
    temp = d.split(" ")
    holder = ""
    if (temp[0] == "Health"):
        holder = temp[0]
    elif (temp[0] == ""):
        holder = "Null"
    elif (temp[0] == "Legal"):
        holder = temp[0]
    elif (len(temp) > 1):
        if (temp[1] == "&"):
            holder = temp[0] + " " + temp[1] + " " + temp[2]
        else:
            holder = temp[0] + " " + temp[1]
    else:
        holder = temp[0]
    if (deptDict.get(holder)!=None):
        if (holder == "Finance, Accounting, & Procurement Policy, Research & Analysis"):
            holder = "Finance, Accounting,"
        holder = holder.strip(',')
        newDepartments.append(holder)
        
```

```python
print(len(newDepartments))
```

```python
df['newDepartments'] = newDepartments
temp = df['Full-Time/Part-Time indicator']
```

```python
counter = 0
while (counter < len(newDepartments)):
    if (len(newDepartments[counter]) <= 2):
        newDepartments[counter] = temp[counter]
    counter +=1
```

```python
newDepartments2 = []
for item in newDepartments:
    if item == "Finance, Accounting, & Procurement Policy, Research & Analysis":
        item = "Finance, Accounting"
        newDepartments2.append(item)
    elif item == "Administration & Human Resources":
        item = "Administration & Human"
        newDepartments2.append(item)
    elif item == "Communications & Intergovernmental Affairs":
        item = "Communications & Intergovernmental"
        newDepartments2.append(item)
    elif item == "Engineering, Architecture, & Planning Policy, Research & Analysis":
        item = "Engineering, Architecture"
        newDepartments2.append(item)
    elif item == "Technology, Data & Innovation":
        item = "Technology, Data"
        newDepartments2.append(item)
    elif item == "Legal Affairs":
        item = "Legal"
        newDepartments2.append(item)
    elif item == "Policy & Analysis":
        item = "Policy"
        newDepartments2.append(item)
    elif item == "Policy, Research":
        item = "Policy"
        newDepartments2.append(item)
    else:
        newDepartments2.append(item)
    print(item)
    newDD[item] = 1
```

print(newDepartments)

```python
df['newDepartments'] = newDepartments2
df.to_csv('updatedJobs.csv')
```
