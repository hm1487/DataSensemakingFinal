```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("NYC_Jobs.csv")
```

```python
departments = df['Job Category']
print(len(departments))
```

```python
deptDict = {}
for d in departments:
    temp = d.split(" ")
    if (temp[0] == "Health"):
        holder = temp[0]
    elif (temp[0] == ""):
        holder = "Null"
    elif (temp[0] == "Legal"):
        holder = temp[0]
    elif (len(temp) > 1):
        if (temp[1] == "&"):
            holder = temp[0] + " " + temp[1] + " " + temp[2]
        else:
            holder = temp[0] + " " + temp[1]
    else:
        holder = temp[0]
    if (deptDict.get(holder)==None):
        deptDict[holder] = 1
    else:
        deptDict[holder] += 1
```

```python
for x in sorted(deptDict, key=deptDict.get, reverse=True):
    if (x == ""):
        print("Null", deptDict.get(x))
    else:
        print(x,deptDict.get(x))
total = 0
for x in deptDict.values():
    total += x
print(total)
```

```python
newDepartments = []
for d in departments:
    temp = d.split(" ")
    holder = ""
    if (temp[0] == "Health"):
        holder = temp[0]
    elif (temp[0] == ""):
        holder = "Null"
    elif (temp[0] == "Legal"):
        holder = temp[0]
    elif (len(temp) > 1):
        if (temp[1] == "&"):
            holder = temp[0] + " " + temp[1] + " " + temp[2]
        else:
            holder = temp[0] + " " + temp[1]
    else:
        holder = temp[0]
    if (deptDict.get(holder)!=None):
        newDepartments.append(holder)
```

```python
print(len(newDepartments))
```

```python
df['newDepartments'] = newDepartments
```

```python
print(df)
```

```python
df.to_csv('updatedJobs.csv')
```
