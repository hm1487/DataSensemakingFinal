```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("NYC_Jobs.csv")
```

```python
departments = df['Job Category']
print(len(departments))
```

```python
deptDict = {}
for d in departments:
    temp = d.split(" ")
    if (temp[0] == "Health"):
        holder = temp[0]
    elif (temp[0] == ""):
        holder = "Null"
    elif (temp[0] == "Legal"):
        holder = temp[0]
    elif (len(temp) > 1):
        if (temp[1] == "&"):
            holder = temp[0] + " " + temp[1] + " " + temp[2]
        else:
            holder = temp[0] + " " + temp[1]
    else:
        holder = temp[0]
    if (deptDict.get(holder)==None):
        deptDict[holder] = 1
    else:
        deptDict[holder] += 1
```

```python
for x in sorted(deptDict, key=deptDict.get, reverse=True):
    if (x == ""):
        print("Null", deptDict.get(x))
    else:
        print(x,deptDict.get(x))
total = 0
for x in deptDict.values():
    total += x
print(total)
```

```python
newDepartments = []
newDD = {}
for d in departments:
    temp = d.split(" ")
    holder = ""
    if (temp[0] == "Health"):
        holder = temp[0]
    elif (temp[0] == ""):
        holder = "Null"
    elif (temp[0] == "Legal"):
        holder = temp[0]
    elif (len(temp) > 1):
        if (temp[1] == "&"):
            holder = temp[0] + " " + temp[1] + " " + temp[2]
        else:
            holder = temp[0] + " " + temp[1]
    else:
        holder = temp[0]
    if (deptDict.get(holder)!=None):
        if (holder == "Finance, Accounting, & Procurement Policy, Research & Analysis"):
            holder = "Finance, Accounting,"
        holder = holder.strip(',')
        newDepartments.append(holder)
        
```

```python
print(len(newDepartments))
```

```python
df['newDepartments'] = newDepartments
temp = df['Full-Time/Part-Time indicator']
```

```python
counter = 0
while (counter < len(newDepartments)):
    if (len(newDepartments[counter]) <= 2):
        newDepartments[counter] = temp[counter]
    counter +=1
```

```python
newDepartments2 = []
for item in newDepartments:
    if item == "Finance, Accounting, & Procurement Policy, Research & Analysis":
        item = "Finance, Accounting"
        newDepartments2.append(item)
    elif item == "Administration & Human Resources":
        item = "Administration & Human"
        newDepartments2.append(item)
    elif item == "Communications & Intergovernmental Affairs":
        item = "Communications & Intergovernmental"
        newDepartments2.append(item)
    elif item == "Engineering, Architecture, & Planning Policy, Research & Analysis":
        item = "Engineering, Architecture"
        newDepartments2.append(item)
    elif item == "Technology, Data & Innovation":
        item = "Technology, Data"
        newDepartments2.append(item)
    elif item == "Legal Affairs":
        item = "Legal"
        newDepartments2.append(item)
    elif item == "Policy & Analysis":
        item = "Policy"
        newDepartments2.append(item)
    elif item == "Policy, Research":
        item = "Policy"
        newDepartments2.append(item)
    else:
        newDepartments2.append(item)
    print(item)
    newDD[item] = 1
```

```python
salaries_from = df['Salary Range From']
salaries_to = df['Salary Range To']
sal_freq = df['Salary Frequency']
sal_freq = sal_freq.copy()
salaries_from = salaries_from.copy()
salaries_to = salaries_to.copy()
print(len(salaries_to))
print(len(salaries_from))
print(len(sal_freq))
```

```python
from math import *
new_sal_from = []
new_sal_to = []

for i in range(len(sal_freq)):
    if sal_freq[i] == "Hourly":
        print("Old from", salaries_from[i])
        print("Old to", salaries_to[i])
        try:
            annual_sal_est_from = float(salaries_from[i]) * 40 * 52 #Approximating annual salary based on hourly salary
            annual_sal_est_to = float(salaries_to[i]) * 40 * 52 #Approximating annual salary based on hourly salary
        except:
            try:
                annual_sal_est_from = float(salaries_from[i+1]) * 40 * 52
                annual_sal_est_to = float(salaries_to[i+2]) * 40 * 52
            except:
                annual_sal_est_from = float(salaries_from[i+2]) * 40 * 52
                annual_sal_est_to = float(salaries_to[i+3]) * 40 * 52
        
        
        print("New from", annual_sal_est_from)
        print("New to", annual_sal_est_to)
        
        new_sal_from.append(annual_sal_est_from)
        new_sal_to.append(annual_sal_est_to)
        sal_freq[i] = "Annual"
    else:
        new_sal_from.append(salaries_from[i])
        new_sal_to.append(salaries_to[i])
```

```python
for sal in new_sal_from:
    if sal == 'F' or '':
        sal = 0
    print(sal)
print(len(new_sal_from))
```

```python
for sal in new_sal_to:
    print(sal)
print(len(new_sal_to))
```

```python
new_sal_diff = []
for i in range(3451):
    try:
        avg_sal = round((float(new_sal_to[i]) + float(new_sal_from[i])) / 2 / 5000.0) * 5000.0
    except:
        avg_sal = round((float(new_sal_to[len(new_sal_diff) % i]) + float(new_sal_from[len(new_sal_diff) % i])) / 2 / 5000.0) * 5000.0
    print(avg_sal)
    new_sal_diff.append(str(avg_sal))

print(len(new_sal_diff))
```
