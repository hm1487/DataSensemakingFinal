```python
import pandas as pd
import numpy as np
import collections
```

```python
df = pd.read_csv("NYC_Jobs.csv")
```

```python
desc = df['Job Description']
skills = df['Preferred Skills']
reqs = df['Minimum Qual Requirements']
print(len(desc))
```

```python
descdict = {}
skillsdict = {}
reqsdict = {}
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words.add('&')
stop_words.add('The')
stop_words.add('nyc')
stop_words.add('new')
stop_words.add('york')
stop_words.add('city')
stop_words.add('')
stop_words.add('â€¢')
stop_words.add('and/or')
stop_words.add('-')
stop_words.add('*')
stop_words.add('â€“')
stop_words.add('+')


```

```python
print("and/or")
```

```python
for description in desc:
    temp = description.split(" ")
    for x in temp:
        x= x.strip(".")
        x= x.strip(";")
        x= x.strip(",")
        x= x.strip("(")
        x= x.strip(")")
        x= x.strip("\'")
        x= x.strip("\\")
        x= x.strip("?")
        x= x.strip("!")
        x= x.strip(">")
        x= x.strip("<")
        x= x.strip("*")
        x= x.strip("-")
        if (x.lower() not in stop_words):
            if (descdict.get(x.lower()) == None):
                descdict[x.lower()] = 1
            else:
                descdict[x.lower()] += 1
```

```python
for x in sorted(descdict, key=descdict.get, reverse=True):
    if (x == ""):
        print("(empty space)",descdict.get(x))
    else:
        print(x,descdict.get(x))
```

```python
for s in skills:
    temp = s.split(" ")
    for x in temp:
        x= x.strip(".")
        x= x.strip(";")
        x= x.strip(",")
        x= x.strip("(")
        x= x.strip(")")
        x= x.strip("\'")
        x= x.strip("\\")
        x= x.strip("?")
        x= x.strip("!")
        x= x.strip(">")
        x= x.strip("<")
        x= x.strip("*")
        x= x.strip("-")
        if (x.lower() not in stop_words):
            if (skillsdict.get(x) == None):
                skillsdict[x.lower()] = 1
            else:
                skillsdict[x.lower()] += 1
```

```python
for x in sorted(skillsdict, key=skillsdict.get, reverse=True):
    if (x == ""):
        print("(empty space)",skillsdict.get(x))
    else:
        print(x,skillsdict.get(x))
```

```python
for r in reqs:
    holder = str(r)
    temp = holder.split(" ")
    for x in temp:
        x= x.strip(".")
        x= x.strip(";")
        x= x.strip(",")
        x= x.strip("(")
        x= x.strip(")")
        x= x.strip("\'")
        x= x.strip("\\")
        x= x.strip("?")
        x= x.strip("!")
        x= x.strip(">")
        x= x.strip("<")
        x= x.strip("*")
        x= x.strip("-")
        x= x.strip("\"")
        if (x.lower() not in stop_words):
            if (reqsdict.get(x) == None):
                reqsdict[x] = 1
            else:
                reqsdict[x] += 1
```

```python
for x in sorted(reqsdict, key=reqsdict.get, reverse=True):
    if (x == ""):
        print("(empty space)",reqsdict.get(x))
    else:
        print(x,reqsdict.get(x))
```
