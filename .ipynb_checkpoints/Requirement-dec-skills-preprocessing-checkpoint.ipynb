```python
import pandas as pd
import numpy as np
import collections
```

```python
df = pd.read_csv("NYC_Jobs.csv")
```

```python
desc = df['Job Description']
skills = df['Preferred Skills']
reqs = df['Minimum Qual Requirements']
```

```python
descdict = {}
skillsdict = {}
reqsdict = {}
```

```python
for description in desc:
    temp = description.split(" ")
    for x in temp:
        if (descdict.get(x) == None):
            descdict[x] = 1
        else:
            descdict[x] += 1
```

```python
for x in sorted(descdict, key=descdict.get, reverse=True):
    if (x == ""):
        print("(empty space)",descdict.get(x))
    else:
        print(x,descdict.get(x))
```

```python
for s in skills:
    temp = s.split(" ")
    for x in temp:
        if (skillsdict.get(x) == None):
            skillsdict[x] = 1
        else:
            skillsdict[x] += 1
```

```python
for x in sorted(skillsdict, key=skillsdict.get, reverse=True):
    if (x == ""):
        print("(empty space)",skillsdict.get(x))
    else:
        print(x,skillsdict.get(x))
```

```python
for r in reqs:
    holder = str(r)
    temp = holder.split(" ")
    for x in temp:
        if (reqsdict.get(x) == None):
            reqsdict[x] = 1
        else:
            reqsdict[x] += 1
```

```python
for x in sorted(reqsdict, key=reqsdict.get, reverse=True):
    if (x == ""):
        print("(empty space)",reqsdict.get(x))
    else:
        print(x,reqsdict.get(x))
```
